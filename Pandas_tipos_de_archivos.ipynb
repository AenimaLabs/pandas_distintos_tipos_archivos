```python
import pandas as pd

# Replace 'ruta/al/archivo.csv' with the actual path to your CSV file in Google Drive
csv_path = '/content/drive/MyDrive/OracleOne/superstore_data_punto_coma.csv'

try:
    # df_csv_punto_coma = pd.read_csv(csv_path, sep=';', nrows=3, usecols=["Id", "Year_Birth", "Income"])
    df_csv_punto_coma = pd.read_csv(csv_path, sep=';', nrows=3, usecols=[0,1,4])
    #nrows es la cantidad de filas que va a leer
    #usecols cuáles columnas voy a traer al df
    #finalmente, las tres primeras filas de las columnas seleccionadas


    print("CSV file imported successfully:")
    display(df_csv_punto_coma.head())
except FileNotFoundError:
    print(f"Error: The file was not found at {csv_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```
